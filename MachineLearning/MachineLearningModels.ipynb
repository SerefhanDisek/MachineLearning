```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# Veri setini yükleyin
data = pd.read_csv("/content/drive/MyDrive/makine öğrenmesi/Grisoni_et_al_2016_EnvInt88.csv")
data.drop(['CAS', 'SMILES', 'Set'], axis=1, inplace=True)

# KNN modeli
x, y = data.loc[:, data.columns != 'Class'], data.loc[:, 'Class']
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x, y)
prediction = knn.predict(x)
print('Prediction: {}'.format(prediction))

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train, y_train)
prediction = knn.predict(x_test)
print('With KNN (K=3) accuracy is: ', knn.score(x_test, y_test))

# Performans metrikleri
accuracy = accuracy_score(y_test, prediction)
recall = recall_score(y_test, prediction, average='macro')
precision = precision_score(y_test, prediction, average='macro')
f1 = f1_score(y_test, prediction, average='macro')

print(f"Accuracy: {accuracy}")
print(f"Recall: {recall}")
print(f"Precision: {precision}")
print(f"F1 Score: {f1}")

# Model complexity
neig = np.arange(1, 25)
train_accuracy = []
test_accuracy = []

# Farklı k değerleri için döngü
for k in neig:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    train_accuracy.append(knn.score(x_train, y_train))
    test_accuracy.append(knn.score(x_test, y_test))

# Grafik çizimi
plt.figure(figsize=[13, 8])
plt.plot(neig, test_accuracy, label='Testing Accuracy')
plt.plot(neig, train_accuracy, label='Training Accuracy')
plt.legend()
plt.title('k-value VS Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(neig)
plt.savefig('graph.png')
plt.show()

print("Best accuracy is {} with K = {}".format(np.max(test_accuracy), 1 + test_accuracy.index(np.max(test_accuracy))))


#%%
##DECISION TREE
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix

# Veri setini yükleyin
data = pd.read_csv("/content/drive/MyDrive/makine öğrenmesi/Grisoni_et_al_2016_EnvInt88.csv")
data.drop(['CAS', 'SMILES', 'Set'], axis=1, inplace=True)

# Veriyi normalleştirme
x_data = data.drop(["Class"], axis=1)
y = data.loc[:, 'Class']
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

# Decision Tree modeli
dt = DecisionTreeClassifier(criterion="entropy", max_depth=3)
dt.fit(x_train, y_train)

# Tahmin ve skor
y_pred = dt.predict(x_test)
print("Accuracy: ", dt.score(x_test, y_test))

# Performans metrikleri
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='macro')
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print(f"Accuracy: {accuracy}")
print(f"Recall: {recall}")
print(f"Precision: {precision}")
print(f"F1 Score: {f1}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)


#%%
#RBF SVM

from sklearn.svm import SVC
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# Veriyi yükleyin ve ön işleme yapın
data = pd.read_csv("/content/drive/MyDrive/makine öğrenmesi/Grisoni_et_al_2016_EnvInt88.csv")
data.drop(['CAS', 'SMILES', 'Set'], axis=1, inplace=True)

x_data = data.drop(["Class"], axis=1)
y = data.loc[:, 'Class']
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))

# Train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

list_accuracy = []
list_recall = []
list_precision = []
list_f1 = []

# Modeli farklı C değerleri ile eğitin ve performans metriklerini hesaplayın
for i in range(1, 100):
    svm = SVC(C=i, random_state=42)
    svm.fit(x_train, y_train)
    y_pred = svm.predict(x_test)

    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred, average='macro')
    precision = precision_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    list_accuracy.append(accuracy)
    list_recall.append(recall)
    list_precision.append(precision)
    list_f1.append(f1)

# En iyi C değerini bulun
best_c_index = np.argmax(list_accuracy)
best_c = best_c_index + 1

print("Support Vector Machine accuracy:", list_accuracy[best_c_index])
print("Best C value:", best_c)
print("Recall:", list_recall[best_c_index])
print("Precision:", list_precision[best_c_index])
print("F1 Score:", list_f1[best_c_index])

# Performans metriklerini grafikte gösterin
plt.figure(figsize=(12, 6))
plt.plot(range(1, 100), list_accuracy, label="Accuracy", color="blue", linestyle=":", marker="*")
plt.plot(range(1, 100), list_recall, label="Recall", color="red", linestyle="--", marker="o")
plt.plot(range(1, 100), list_precision, label="Precision", color="green", linestyle="-.", marker="s")
plt.plot(range(1, 100), list_f1, label="F1 Score", color="purple", linestyle="-", marker="d")
plt.xlabel("C Values")
plt.ylabel("Scores")
plt.legend()
plt.show()


#%%
# Naive Bayes
from sklearn.naive_bayes import BernoulliNB
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# Veriyi yükleyin ve ön işleme yapın
data = pd.read_csv("/content/drive/MyDrive/makine öğrenmesi/Grisoni_et_al_2016_EnvInt88.csv")
data.drop(['CAS', 'SMILES', 'Set'], axis=1, inplace=True)

x_data = data.drop(["Class"], axis=1)
y = data.loc[:, 'Class']
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))

# Train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

# Modeli eğitin ve tahmin yapın
nb = BernoulliNB()
nb.fit(x_train, y_train)

y_pred = nb.predict(x_test)

# Performans metriklerini hesaplayın
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='macro')
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("Naive Bayes Bernoulli accuracy: ", accuracy)
print("Recall: ", recall)
print("Precision: ", precision)
print("F1 Score: ", f1)


#LSVM
from sklearn.svm import LinearSVC
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# Veriyi yükleyin ve ön işleme yapın
data = pd.read_csv("/content/drive/MyDrive/makine öğrenmesi/Grisoni_et_al_2016_EnvInt88.csv")
data.drop(['CAS', 'SMILES', 'Set'], axis=1, inplace=True)

x_data = data.drop(["Class"], axis=1)
y = data.loc[:, 'Class']
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))

# Train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

# Değer aralığı belirleme
c_values = np.linspace(0.1, 1.0, 10)

list_accuracy = []
list_recall = []
list_precision = []
list_f1 = []

# Modeli farklı C değerleri ile eğitin ve performans metriklerini hesaplayın
for c in c_values:
    svm = LinearSVC(C=c, random_state=42)
    svm.fit(x_train, y_train)
    y_pred = svm.predict(x_test)

    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred, average='macro')
    precision = precision_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    list_accuracy.append(accuracy)
    list_recall.append(recall)
    list_precision.append(precision)
    list_f1.append(f1)

# En iyi C değerini bulun
best_c_index = np.argmax(list_accuracy)
best_c = c_values[best_c_index]

print("Linear Support Vector Machine accuracy:", list_accuracy[best_c_index])
print("Best C value:", best_c)
print("Recall:", list_recall[best_c_index])
print("Precision:", list_precision[best_c_index])
print("F1 Score:", list_f1[best_c_index])

# Performans metriklerini grafikte gösterin
plt.figure(figsize=(12, 6))
plt.plot(c_values, list_accuracy, label="Accuracy", color="blue", linestyle=":", marker="*")
plt.plot(c_values, list_recall, label="Recall", color="red", linestyle="--", marker="o")
plt.plot(c_values, list_precision, label="Precision", color="green", linestyle="-.", marker="s")
plt.plot(c_values, list_f1, label="F1 Score", color="purple", linestyle="-", marker="d")
plt.xlabel("C Values")
plt.ylabel("Scores")
plt.legend()
plt.show()





#%%
# RANDOM FOREST
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Veriyi yükleyin ve ön işleme yapın
data = pd.read_csv("/content/drive/MyDrive/makine öğrenmesi/Grisoni_et_al_2016_EnvInt88.csv")
data.drop(['CAS', 'SMILES', 'Set'], axis=1, inplace=True)

x_data = data.drop(["Class"], axis=1)
y = data.loc[:, 'Class']
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

# Modeli eğitin ve tahmin yapın
rf = RandomForestClassifier(n_estimators=99, random_state=42)  # n_estimators değeri 99 olarak seçildi
rf.fit(x_train, y_train)

y_pred = rf.predict(x_test)

# Performans metriklerini hesaplayın
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='macro')
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("Random Forest accuracy: ", accuracy)
print("Recall: ", recall)
print("Precision: ", precision)
print("F1 Score: ", f1)

# Farklı n_estimators değerleri için accuracy hesaplama ve plot
list_accuracy = []
for i in range(1, 100):
    rf = RandomForestClassifier(n_estimators=i, random_state=42)
    rf.fit(x_train, y_train)
    list_accuracy.append(rf.score(x_test, y_test))

plt.figure(figsize=(12, 6))
plt.plot(range(1, 100), list_accuracy,
         color="blue",
         linewidth=3,
         linestyle=":",
         marker="o",
         markerfacecolor="yellow",
         markeredgewidth=2,
         markeredgecolor="blue")
plt.xlabel("n_estimators Values")
plt.ylabel("Accuracy")
plt.show()

print("Random Forest algo result with n_estimators=99: ", rf.score(x_test, y_test))


# MLP Classifier
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier

# Veriyi yükleyin ve ön işleme yapın
data = pd.read_csv("/content/drive/MyDrive/makine öğrenmesi/Grisoni_et_al_2016_EnvInt88.csv")
data.drop(['CAS', 'SMILES', 'Set'], axis=1, inplace=True)

x_data = data.drop(["Class"], axis=1)
y = data.loc[:, 'Class']
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

# Veriyi normalize edin
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# MLP modelini oluşturun ve eğitin
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42)
mlp.fit(x_train, y_train)

# Tahmin yapın
y_pred = mlp.predict(x_test)

# Performans metriklerini hesaplayın
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='macro')
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("MLP Classifier accuracy: ", accuracy)
print("Recall: ", recall)
print("Precision: ", precision)
print("F1 Score: ", f1)

# Eğitim ve doğrulama kayıplarını görselleştirme
plt.figure(figsize=(12, 6))
plt.plot(mlp.loss_curve_, label='Training Loss')
plt.title('MLP Classifier Training Loss')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.show()


#Logistic Regression
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# Veri setini yükleyin ve gereksiz sütunları kaldırın
data = pd.read_csv("/content/drive/MyDrive/makine öğrenmesi/Grisoni_et_al_2016_EnvInt88.csv")
data.drop(['CAS', 'SMILES', 'Set'], axis=1, inplace=True)

# Bağımsız ve bağımlı değişkenleri ayırın
x = data.drop(["Class"], axis=1)
y = data["Class"]

# Veri setini eğitim ve test setlerine ayırın
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

# Sınıf etiketlerini dönüştürme
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Logistic Regression modelini oluşturun ve eğitin
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(x_train, y_train_encoded)

# Test seti üzerinde tahmin yapın
y_pred = log_reg.predict(x_test)

# Modelin performansını değerlendirin
accuracy = accuracy_score(y_test_encoded, y_pred)
recall = recall_score(y_test_encoded, y_pred, average='macro')
precision = precision_score(y_test_encoded, y_pred, average='macro')
f1 = f1_score(y_test_encoded, y_pred, average='macro')

# Sonuçları yazdırın
print("Logistic Regression accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
print("F1 Score:", f1)



#Neural Networks
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# Veri setini yükleyin ve gereksiz sütunları kaldırın
data = pd.read_csv("/content/drive/MyDrive/makine öğrenmesi/Grisoni_et_al_2016_EnvInt88.csv")
data.drop(['CAS', 'SMILES', 'Set'], axis=1, inplace=True)

# Bağımsız ve bağımlı değişkenleri ayırın
x = data.drop(["Class"], axis=1)
y = data["Class"]

# Veri setini eğitim ve test setlerine ayırın
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

# Sınıf etiketlerini dönüştürme
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Yapay Sinir Ağı (Neural Network) modelini oluşturun ve eğitin
mlp = MLPClassifier(hidden_layer_sizes=(100, 50), activation='relu', solver='adam', max_iter=1000, random_state=42)
mlp.fit(x_train, y_train_encoded)

# Test seti üzerinde tahmin yapın
y_pred = mlp.predict(x_test)

# Modelin performansını değerlendirin
accuracy = accuracy_score(y_test_encoded, y_pred)
recall = recall_score(y_test_encoded, y_pred, average='macro')
precision = precision_score(y_test_encoded, y_pred, average='macro')
f1 = f1_score(y_test_encoded, y_pred, average='macro')

# Sonuçları yazdırın
print("Neural Network accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
print("F1 Score:", f1)



#AdaBoost
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# Veri setini yükleyin ve gereksiz sütunları kaldırın
data = pd.read_csv("/content/drive/MyDrive/makine öğrenmesi/Grisoni_et_al_2016_EnvInt88.csv")
data.drop(['CAS', 'SMILES', 'Set'], axis=1, inplace=True)

# Bağımsız ve bağımlı değişkenleri ayırın
x = data.drop(["Class"], axis=1)
y = data["Class"]

# Veri setini eğitim ve test setlerine ayırın
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

# AdaBoost modelini oluşturun ve eğitin
adaboost = AdaBoostClassifier(n_estimators=50, random_state=42)
adaboost.fit(x_train, y_train)

# Test seti üzerinde tahmin yapın
y_pred = adaboost.predict(x_test)

# Modelin performansını değerlendirin
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='macro')
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

# Sonuçları yazdırın
print("AdaBoost Classifier accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
print("F1 Score:", f1)





```
